```python
names = ["Donald Duck", "Mickey Mouse", "Daffy Duck", "Goofy", "Minnie Mouse", "Pluto"]
```

```python
def get_total_length(names):
    # Initialize result
    total_length = 0

    # Loop through data
    for name in names:

        # Modify result
        total_length += len(name)

    # Show the result
    return total_length
```

```python
def calc_profit(quantity, hours):

    lemon_cost = 0.5
    lemons_needed = 4
    
    # Compute unit cost from number of lemons per lemonade and per-lemon cost
    unit_cost = lemon_cost * lemons_needed
    
    # Revenue per lemonade is given as $5
    unit_revenue = 5

    # Compute unit profit as unit revenue - unit cost
    unit_profit = unit_revenue - unit_cost

    # Multiply by quantity to get total profit
    total_profit = unit_profit * quantity

    # Divide by hours worked to get hourly profit
    hourly_profit = total_profit / hours

    # Return result as an interpolated string
    return f"You made ${total_profit}, averaging ${hourly_profit} per hour."
    
    
```

```python
calc_profit(10, 2)
```

```python
def calc_profit(quantity, hours, price=5, lemon_cost=0.5, lemons_needed=4):
    profit = quantity * (price - lemon_cost * lemons_needed)
    return f"You made ${profit}, averaging ${profit / hours} per hour."
    
```

```python
calc_profit(10, 2, price=7, lemon_cost=0.75)
```

```python
def get_aggregate_order_counts(orders):
    
    # initialize
    tally = {}
    
    for order in orders:
        
        # if seen before, increment count
        if order in tally:
            tally[order] += 1

        # else set count to 1
        else:
            tally[order] = 1
        
    return tally
```

```python
freqs = get_aggregate_order_counts(["burger", "fries", "burger", "tenders", "apple pie"])
```

```python
def is_valid_year(song_release_year):
    try:
        year = int(song_release_year)
    except ValueError:
        return False
    else:
        return year > 1900
```
